```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('02 real estate price size.csv')
```

```python
data
```

```python
data.describe()
```

```python
#Create regression
```

```python
#Define dependent and independent variables
```

```python
y = data['size']
x1 = data['price']
```

```python
#explore the data
```

```python
plt.scatter(x1,y)
plt.xlabel('price', fontsize = 10)
plt.ylabel('size', fontsize = 10)
plt.show()
```

```python
#Regression here
#use coeff table to calculate regression formula
#std error > accuracy
#t-stat and P-value
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat = 0.0033 * x1 - 122.3349
fig = plt.plot(x1, yhat, lw=4, c='orange', label='regression line')
plt.xlabel('price', fontsize = 10)
plt.ylabel('size', fontsize = 10)
```
