```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
#Reorder the columns
```

```python
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df = df[column_names_reordered]
```

```python
# working with date feature
```

```python
type(df['Date'][0])
```

```python
df_datetime = df.copy()
```

```python
df_datetime['Date'] = pd.to_datetime(df_datetime['Date'], format = '%d/%m/%Y')
```

```python
df_datetime['Date']
```

```python
df_datetime.info()
```

```python
#Extract month values
```

```python
list_month = []
```

```python
for i in range(700):
  list_month.append(df_datetime['Date'][i].month)
```

```python
df_datetime['Month Value'] = list_month
```

```python
df_datetime.head()
```

```python
#Extract day of the week
```

```python
def date_to_weekday(date_value):
  return date_value.weekday()
```

```python
df_datetime['Day of the week'] = df_datetime['Date'].apply(date_to_weekday)
```

```python
df_datetime.head()
```

```python
#Remove date column
```

```python
df_nodate = df_datetime.drop(['Date'], axis = 1)
```

```python
df_nodate.columns.values
```

```python
rearrange_feature = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value',
       'Day of the week', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df_preprocessed = df_nodate[rearrange_feature]
```

```python
df_preprocessed.head(10)
```

```python
df_preprocessed['Education'].value_counts()
```

```python
# Education 1 is high school, 2 is graduate, 3 is postgraduate, 4 is master or doctor
```

```python
df_preprocessed['Education'] = df_preprocessed['Education'].map({1:0, 2:1, 3:1, 4:1})
```
