```python
import numpy as np
import tensorflow as tf
```

```python
#Load the data
```

```python
npz = np.load('/content/drive/MyDrive/Colab Notebooks/Deep Learning/Audiobooks_data_train.npz')

train_inputs = npz['inputs'].astype(np.float)
train_targets = npz['targets'].astype(np.int)

npz = np.load('/content/drive/MyDrive/Colab Notebooks/Deep Learning/Audiobooks_data_validation.npz')

validation_inputs, validation_targets = npz['inputs'].astype(np.float), npz['targets'].astype(np.int)
```

```python
# Model
# --> Outline, optimizers, loss, early stopping and training
```

```python
input_size = 10
output_size = 2
hidden_layer_size = 50

#Model
model = tf.keras.Sequential([
                            tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
                            tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
                            tf.keras.layers.Dense(output_size, activation='softmax')
                            ])

#loss function
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

batch_size = 100

max_epochs = 100

early_stopping = tf.keras.callbacks.EarlyStopping(patience=2)

model.fit(train_inputs,
          train_targets,
          batch_size = batch_size,
          epochs = max_epochs,
          callbacks = [early_stopping],
          validation_data=(validation_inputs, validation_targets),
          verbose = 2)
```

```python
#Test The Model
```

```python
npz = np.load('/content/drive/MyDrive/Colab Notebooks/Deep Learning/Audiobooks_data_test.npz')

test_inputs = npz['inputs'].astype(np.float)
test_targets = npz['targets'].astype(np.int)

test_lost , test_accuracy = model.evaluate(test_inputs, test_targets)
```
