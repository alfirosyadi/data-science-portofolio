```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('05 Dummies.csv')
```

```python
raw_data
```

```python
#imitating dummies variables with numbers (0 and 1)
```

```python
data = raw_data.copy()
```

```python
data['Attendance'] = data['Attendance'].map({'Yes':1,'No':0})
```

```python
data
```

```python
data.describe()
```

```python
##Regression
```

```python
y = data['GPA']
x1 = data[['SAT', 'Attendance']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```

```python
## the R-squared increase with the addition of attendance
```

```python
## the linear regression formula GPA = 0.6439 + 0.0014 * SAT + 0.2226 * Attendance
```

```python
plt.scatter(data['SAT'],y, c=data['Attendance'], cmap='RdYlGn_r')
yhat_no = 0.6439 + 0.0014 * data['SAT']
yhat_yes = 0.8665 + 0.0014 * data['SAT']
fig = plt.plot(data['SAT'], yhat_no, lw=2, c='#006837')
fig = plt.plot(data['SAT'], yhat_yes, lw=2, c='#ffff00')
plt.xlabel('SAT', fontsize=10)
plt.ylabel('GPA', fontsize=10)
plt.show()
```
